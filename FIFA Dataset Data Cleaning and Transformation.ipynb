# Data Cleaning and Manipulation

```python
#remove unnecessary column
Fifa_df = Fifa_df.drop(["photoUrl","playerUrl"], axis = 1)
```

```python
#check duplicate rows
duplicate_rows = Fifa_df[Fifa_df.duplicated(subset=["LongName", "Name"])]
duplicate_rows
```

```python
#remove duplicate rows
Fifa_df = Fifa_df[~Fifa_df.duplicated(subset=["LongName","Name"])]
```

```python
Fifa_df.info()
```

```python
Fifa_df["Positions"].head(60)
```

```python
#split position using split function
position_tokens = Fifa_df["Positions"].str.split()
position_tokens
```

```python
#make empty set to store unique positions
positions = set()
#update each unique positions in set
for position in position_tokens:
    positions.update(position)
positions
```

```python
# Make new dataframe to store encoded values
encoded = pd.DataFrame()
```

```python
#one hot encoded
for token in positions:
    encoded[token] = Fifa_df["Positions"].str.contains(token, regex = False).astype(int)
encoded[token]
```

```python
# Creating function to convert string to numeric
def convert_to_numeric(value_str):
    #Remove currency symbol
    value_str = value_str.replace('€', '').replace(',', '')
    if 'K' in value_str:
        return float(value_str.replace('K', '')) * 1000
    elif 'M' in value_str:
        return float(value_str.replace('M', '')) * 1000000
    else:
        return float(value_str)
# Apply the conversion function to the "Wage" and "Value" columns
Fifa_df["Wage"] = Fifa_df["Wage"].apply(convert_to_numeric)
Fifa_df["Value"] = Fifa_df["Value"].apply(convert_to_numeric)
Fifa_df[["Value", "Wage"]]
```

```python
#apply convert_to_numeric function
Fifa_df["Release Clause"] = Fifa_df["Release Clause"].apply(convert_to_numeric)
Fifa_df["Release Clause"]
```

# Store into excel

```python
excel_file = "combined_data.xlsx"
# Create an ExcelWriter object
with pd.ExcelWriter(excel_file, engine='xlsxwriter') as writer:
    # Save each DataFrame to a separate sheet
    Fifa_df.to_excel(writer, sheet_name="Sheet1", index=False)
    encoded.to_excel(writer, sheet_name="Sheet2", index=False)
```

In this project, we will walk through the process of cleaning and transforming a FIFA dataset. The objective is to prepare the data for analysis by removing inconsistencies and converting it into a more usable format. The following key tasks were undertaken:

## Task 1: Removing URLs

We started by removing URLs from the dataset, ensuring that hyperlinks or website references do not interfere with the analysis.

## Task 2: Removing Duplicates

Duplicate records were identified and removed to maintain data integrity and avoid redundancy.

## Task 3: Converting Positions into One-Hot Encoding

The "Positions" column was converted into one-hot encoding to facilitate further analysis by representing player positions as binary values.

## Task 4: Separating "Team & Contract" into Multiple Columns

We split the "Team & Contract" column into three distinct columns: "Team," "Contract Start Year," and "Contract End Year" for better data organization.

## Task 5: Converting Data Types of Height and Weight

The data types of the "Height" and "Weight" columns were converted, and heights were transformed from feet (ft) to centimeters (cm), while weights were converted from pounds (lbs) to kilograms (kg).

## Task 6: Converting Date Strings into Datetime Format

We converted date strings into the datetime data type, making it easier to work with time-related information. Additionally, we extracted and stored the month, year, and day in separate columns for more detailed analysis.

## Task 7: Converting String into Numeric Money Values

Monetary values within the dataset were converted from strings into numeric formats to enable financial calculations.

## Task 8: Checking for Missing Values

We performed a comprehensive check for missing values within the dataset and filled them as needed to ensure data completeness.

## Saving the Cleaned Data

The cleaned dataset was saved to an Excel workbook with two separate sheets. The first sheet contains the cleaned data, and the second sheet contains the one-hot encoded player positions.
